```python
import copy
class Puzzle:
    def __init__(self,i,g):
        self.current_state = i
        self.goal_state = g
        self.prev_state = None
    def blank_tile(self):
        return self.current_state.index(' ')
    
    def up(self):
        pos = self.blank_tile()
        if pos in [0,1,2]:
            #print("WRONG MOVE")
            return False
        else:
            self.current_state[pos] , self.current_state[pos-3] = self.current_state[pos-3] , self.current_state[pos]
            #self.print_board()
            #self.check_goal_state()
            return True
    def down(self):
        pos = self.blank_tile()
        if pos in [6,7,8]:
            #print("WRONG MOVE")
            return False
        else:
            self.current_state[pos] , self.current_state[pos+3] = self.current_state[pos+3] , self.current_state[pos]
            #self.print_board()
            #self.check_goal_state()
            return True
    def left(self):
        pos = self.blank_tile()
        if pos in [0,3,6]:
            #print("WRONG MOVE")
            return False
        else:
            self.current_state[pos] , self.current_state[pos-1] = self.current_state[pos-1] , self.current_state[pos]
            #self.print_board()
            #self.check_goal_state()
            return True
    def right(self):
        pos = self.blank_tile()
        if pos in [2,5,8]:
            #print("WRONG MOVE")
            return False
        else:
            self.current_state[pos] , self.current_state[pos+1] = self.current_state[pos+1] , self.current_state[pos]
            #self.print_board()
            #self.check_goal_state()
            return True
    def check_goal_state(self):
        return self.current_state == self.goal_state
            #print("GOAL STATE REACHED")
    def print_board(self):
        print('------------')
        print('|   |   |   |')
        print('| ' + self.current_state[0] + ' | ' + self.current_state[1] + ' | ' + self.current_state[2]+ ' | ')
        print('|   |   |   |')
        print('------------')
        print('|   |   |   |')
        print('| ' + self.current_state[3] + ' | ' + self.current_state[4] + ' | ' + self.current_state[5]+ ' | ')
        print('|   |   |   |')
        print('------------')
        print('|   |   |   |')
        print('| ' + self.current_state[6] + ' | ' + self.current_state[7] + ' | ' + self.current_state[8]+ ' | ')
        print('|   |   |   |')
        print('------------')
        print()

def check_closed(n):
    for c in closed:
        if c.current_state==n.current_state:
            return True
    return False

def print_path(c):
    c.print_board()
    print('\t^\n\t:\n\t:\n\t:\n\t:\n\t:\n\n')
    while(c.prev_state != None):
        
        c = c.prev_state
        c.print_board()
        print('\t^\n\t|\n\t|\n\t|\n\t|\n\t|\n\n')

def bfs(in1):
    openl.append(in1)
    while openl!=[]:
        current = openl.pop(0)
        closed.append(current)
        
        if(current.check_goal_state()):
            print_path(current)
            break
        
        
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.left() and check_closed(new_state)==False:
 
            openl.append(new_state)
           # new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.right() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.up() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.down() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
    
def dfs(in1):
    openl.append(in1)
    while openl!=[]:
        current = openl.pop()
        closed.append(current)
        
        if(current.check_goal_state()):
            print_path(current)
            break
        
        
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.up() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.down() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.right() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
            
        new_state = copy.deepcopy(current)
        new_state.prev_state = copy.deepcopy(current)
        if new_state.left() and check_closed(new_state)==False:
            openl.append(new_state)
            #new_state.print_board()
        else:
            del new_state
    
i = ['2',' ','3','1','8','4','7','6','5']
g = ['1','2','3','8',' ','4','7','6','5']
in1 = Puzzle(i,g)
#in1.print_board()
openl = []
closed = []
#dfs(in1)
bfs(in1)
```
